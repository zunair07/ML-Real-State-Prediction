# Removing Outliers in data

```python
from scipy import stats
z = np.abs(stats.zscore(df))
z
```

```python
threshold=3
np.where(z>3)
```

```python
z[56][1]
```

```python
Q1 = df.quantile(0.20)
Q3 = df.quantile(0.80) 
IQR = Q3 - Q1
IQR
```

```python
#TRUE- Shows presence of outliers
(df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))
```

```python
df_new =df[(z < 3).all(axis=1)]
df_new.shape
```

```python
~(df_new < (Q1 - 1.5 * IQR)) | ~(df_new > (Q3 + 1.5 * IQR))
```

```python
df_out = df_new[~((df_new < (Q1 - 1.5 * IQR)) |(df_new > (Q3 + 1.5 * IQR))).any(axis=1)]
```
